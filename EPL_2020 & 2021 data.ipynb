# Analysis EPL_20_21 dataset

```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#loading Data
fb= pd.read_csv('Desktop/EPL_20_21.csv')
```

#### There is no null vakue so we can go for next step for analysing process

```python
#create 2 new columns MinsPerMatch & GoalPer Match
fb['MinsPerMatch']=(fb['Mins'] / fb['Matches'].astype(int))

fb['GoalPerMatch']=(fb['Goals'] / fb['Matches'].astype(float))
```

```python
#print Data with updated columns
fb.head()
```

```python
#Total Goals
total_goals=fb['Goals'].sum()
```

```python
print(total_goals)
```

```python
#counting penailty goals
total_penailty=fb['Penalty_Goals'].sum()
```

```python
print(total_penailty)
```

```python
#Total Penailty Attempt
total_PA = fb['Penalty_Attempted'].sum()
print(total_PA)
```

```python
#pie chart for penalties missed vs scored
plt.figure(figsize=(7,7))
pl_not_scored=fb['Penalty_Attempted'].sum() - total_penailty
data=[pl_not_scored, total_penailty]
labels= ['Penalties missed','Penalties Scored']
color=sns.color_palette('Set2')
plt.pie(data, labels=labels, colors=color, autopct= '%0.2f%%')
plt.show()
```

```python
# Most Player From Which Countries
nationality = fb.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind='bar', figsize=(12,6), color=sns.color_palette('magma'))
plt.show()
```

```python
# clubs with maximum plyarers in their squad
fb['Club'].value_counts()[0:5].plot(kind='bar', color=sns.color_palette("viridis"))
plt.show()
```

```python
#second Method
fb['Club'].value_counts().nlargest(5).plot(kind='bar', color=sns.color_palette("viridis"))
plt.show()
```

```python
#club with least players in their squad
fb['Club'].value_counts().nsmallest(5).plot(kind='bar', color=sns.color_palette("viridis"))
plt.show()
```

```python
#Players Based on age group
under_20 = fb[fb['Age'] <=20]
age20_25 = fb[fb['Age'] > (20 & fb['Age'] <=25)]
age25_30 = fb[fb['Age'] > (25 & fb['Age'] <=30)]
above30 = fb[fb['Age']>30]
```

```python
#making Pie chart
x = np.array([under_20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),above30['Name'].count()])
mylables = ['<=20', '>20 & <=25','>25 & <=30', '>30']
plt.figure(figsize=(12,7))
plt.title('Total players with age groups', fontsize=20)
plt.pie(x, labels=mylables, autopct='%0.2f%%')
plt.show()
```

```python
#total under 20 players in each club
players_under_20 = fb[fb['Age']<20]
players_under_20['Club'].value_counts().plot(kind='bar', color= sns.color_palette('cubehelix'))
plt.show()
```

```python
# under 20 players in Manchesters United
players_under_20[players_under_20['Club']=='Manchester United']
```

```python
# under 20 players in chelsea
players_under_20[players_under_20['Club']=='Chelsea']
```

```python
# Average age of a players in each club
plt.figure(figsize=(12,6))
sns.boxplot(x='Club', y='Age', data=fb)
plt.xticks(rotation=90)     #rotating text in graphic
plt.show()
```

```python
num_player = fb.groupby('Club').size()
data = (fb.groupby('Club')['Age'].sum())/num_player
data.sort_values(ascending = False)
```

```python
#Total Assists from each club

Assits_by_club = pd.DataFrame(fb.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club',y='Assists',data=Assits_by_club.sort_values(by='Assists'),palette='Set2')
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Assists',fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']=(20,8)
plt.title('Plot of club vs total assists', fontsize=20)
plt.show()
```

```python
#top 10 Assists

top_10_assists = fb[['Name','Club','Assists','Matches']].nlargest(n=10, columns='Assists')
top_10_assists
```

```python
# Total Goals From each Club
#Total Assists from each club

Goals_by_club = pd.DataFrame(fb.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club',y='Goals',data=Goals_by_club.sort_values(by='Goals'),palette='rocket')
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Goals',fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']=(20,8)
plt.title('Plot of club vs total Goals', fontsize=20)
plt.show()
```

```python
#Most Goal BY Players

top_10_Goals = fb[['Name','Club','Goals','Matches']].nlargest(n=10, columns='Goals')
top_10_Goals
```

```python
#Goals Per match

top_10_Goals_per_match = fb[['Name','GoalPerMatch','Matches','Goals']].nlargest(n=10, columns='GoalPerMatch')
top_10_Goals_per_match
```

```python
#pie chart  - Goals with assist and without assist

plt.figure(figsize=(12,6))
assists=fb['Assists'].sum()
data = [total_goals - assists, assists]
labels=['Goals Without assists', 'Goals With assists']
color = sns.color_palette('Set2')
plt.pie(data, labels=labels, colors = color, autopct="%0.2f%%")
plt.show()
```

```python
#Top 10 players with most yellow cards

fb_yellow = fb.sort_values(by='Yellow_Cards', ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title('playrer with the most yellow card')
c = sns.barplot(x=fb_yellow['Name'], y=fb_yellow['Yellow_Cards'], label='Players', color='Yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(), rotation=45)
plt.show()
```
